# Preference aggregation

Single preference is a [binary relation](https://en.wikipedia.org/wiki/Binary_relation) between two alternatives.

**Preference ordering**: ranking of a set of alternatives (usually within some category) - e.g. (A)pples > (B)ananas > (C)oconuts
* Revealed through actions (e.g. as in [Revealed Preference Theory](https://en.wikipedia.org/wiki/Revealed_preference))

Preference orderings are a different mathematical structure than the objects we were aggregating before (numbers/actions and rules) - they are a [relation](https://en.wikipedia.org/wiki/Binary_relation), more specifically - a [preorder](https://en.wikipedia.org/wiki/Preorder).

**How many rational preference orderings there may be?**



For 3 alternatives, we have $3 \times 2 \times 1 = 3!$ rational preference orderings. In general these are [permutations](https://en.wikipedia.org/wiki/Permutation), and for $n$ alternatives there are $n!$ rational (transitive) preference orderings.

More elaborate models can also allow an equality of alternatives.

**How to aggregate preference orderings?**

E.g., how to aggregate preferences of a set of people? (The problem also appears in other domains, i.e. aggregating rankings in general.)

Let's see how a set of pairwise preferences (votes) play out:



What is the collective preference ordering according to these votes?



This is an **irrational** (or intransitive) set of aggregated preferences, even though all the individuals had a set of rational preferences! Expecting otherwise is called a ["fallacy of composition"](https://en.wikipedia.org/wiki/Fallacy_of_composition).

This paradox is called the [Condorcet Paradox](https://en.wikipedia.org/wiki/Condorcet_paradox) (or Voting Paradox, an important result in [social choice theory](https://en.wikipedia.org/wiki/Social_choice_theory)):
* Each person's preferences rational (transitive)
* Collective aggregated preferences *not* rational (transitive)

See also: [Arrow's impossibility theorem](https://en.wikipedia.org/wiki/Arrow%27s_impossibility_theorem)

The fact that there is a problem with aggregation of preferences, can cause strategic voting (to be examined later in course).
