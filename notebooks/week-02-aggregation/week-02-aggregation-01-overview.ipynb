# Aggregation (overview)

Logic of aggregation can be surprising, which is why it's worth modeling!


* "More is different" - [1972 paper](https://doi.org/10.1126/science.177.4047.393) by [Philip W. Anderson](https://en.wikipedia.org/wiki/Philip_W._Anderson) (1977 Nobel prize in physics)
    * Opposed to naive reductionism
    * Ex: molecule of H2O is not wet, but water is
    * Ex: *emergent properties* of the brain

Models of **aggregation** that will be considered, and what they enable:
* Independent actions - Central Limit Theorem
    * Predict points
    * Understand data
* Single rule - Game of Life
    * Understand patterns
* Family of rules - 1D Cellular Automata
    * Understand classes of outcomes
* Preferences
    * Work through logic
    
How to understand how systems work (e.g. social systems)?
* Understand how parts work (e.g. individual decision making - next module)
* Understand how parts add up (i.e. aggregation - this module)
