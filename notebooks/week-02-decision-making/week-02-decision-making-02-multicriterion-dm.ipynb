# Multi-Criterion Decision Making

Multi-Criterion Decision Making (aka [Multiple-criteria decision analysis](https://en.wikipedia.org/wiki/Multiple-criteria_decision_analysis), also related to [Multi-objective optimization](https://en.wikipedia.org/wiki/Multi-objective_optimization)) involves deciding between different choices (solutions/alternatives), based on a number of criteria (dimensions/attributes/features). The analysis can be informal (qualitative) or formal (quantitative). The goal is to identify the best choice, or a set of [non-dominated / non-inferior](Maxima_of_a_point_set) choices (see [Pareto efficiency / optimality](https://en.wikipedia.org/wiki/Pareto_efficiency)).
